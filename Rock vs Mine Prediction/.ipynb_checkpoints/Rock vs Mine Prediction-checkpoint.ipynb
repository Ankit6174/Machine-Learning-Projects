# Rock vs Mine Prediction Wirh Logistic Regression

```python
# First'o fall we want to load our dataset 

import pandas as pd

data = pd.read_csv('./Copy of sonar data.csv')
```

```python
data.head() ## Printing first 5 rows of our dataset
```

```python
data['R'] = data['R'].map({'R': 'Rock', 'M': 'Mine'})  # Converting our targeted column into a perfect manner
```

```python
data.head()
```

```python
data.shape   # Rows and columns in our dataset
```

```python
x = data.iloc[:, :-1]  # Selecting our features and target column
y = data.iloc[:, -1]
```

## Splitting data into test & train

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=12)
```

## Train our model with the help of training values.

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(x_train, y_train)
```

### Predicting values

```python
y_pred = model.predict(x_test)
```

## Accuracy score of training and testing data.

```python
from sklearn.metrics import accuracy_score

x_train_prediction = model.predict(x_train)
accuracy_Score = accuracy_score(x_train_prediction, y_train)
print(f'On training data our is {accuracy_Score*100}% correct 👍.')

x_test_prediction = model.predict(x_test)
accuracy_Score = accuracy_score(x_test_prediction, y_test)
print(f'On testing data our is {accuracy_Score*100}% correct 👍.')
```
